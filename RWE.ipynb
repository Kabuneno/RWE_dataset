```python
#фунцкия для обработки средней степени так скажем
for i in range(0,len(sentences)):
  midsentences[i] = midsentences[i].casefold()
  midpredict[i]= midpredict[i].casefold()
  midsentences[i] = re.sub(r'[^\w\s]', '', midsentences[i])
  midpredict[i] = re.sub(r'[^\w\s]', '', midpredict[i])
print(midpredict[0])

# finalssent = midsentences.copy()
# finalpredict = midpredict.copy()
# #функция для истинного вычисления совпадений высокая степень обработки
# for i in range(0,len(sentences)):
#   finalssent[i] = finalssent[i].replace(" ","")
#   finalpredict[i] = finalpredict[i].replace(" ","")
```
